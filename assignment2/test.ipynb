```python
import pandas as pd

data = {
    'Car Age': [5, 3, 8, 2, 6, 4, 7, 1, 9],
    'Previous Owners': [1, 2, 3, 1, 2, 1, 3, 1, 4],
    'Price': [20, 25, 15, 30, 18, 22, 16, 35, 12]
}

def calculate_spearman(x, y):
    # 计算秩次（处理重复值）
    x_rank = x.rank(method='average')
    y_rank = y.rank(method='average')
    
    # 计算秩差平方和
    d_sq = (x_rank - y_rank)**2
    sum_d_sq = d_sq.sum()
    
    # 斯皮尔曼系数公式
    n = len(x)
    rho = 1 - (6 * sum_d_sq) / (n * (n**2 - 1))
    return rho

# 提取数据
car_age = pd.Series(data['Car Age'])
owners = pd.Series(data['Previous Owners'])
price = pd.Series(data['Price'])

# 验证秩次计算
owners_rank = owners.rank(method='average').round(2)
price_rank = price.rank(ascending=False, method='average').round(2)  # Price是价格越高排名越低

print("Previous Owners 秩次:", list(owners_rank))
print("Price 秩次:", list(price_rank))

# 计算相关系数
car_age_price = calculate_spearman(car_age, price)
owners_price = calculate_spearman(owners, price)

print(f"\nCar Age-Price Spearman ρ: {car_age_price:.3f}")
print(f"Previous Owners-Price Spearman ρ: {owners_price:.3f}")
```

```python


data = {
    'Car Age': [5, 3, 8, 2, 6, 4, 7, 1, 9],
    'Previous Owners': [1, 2, 3, 1, 2, 1, 3, 1, 4],
    'Price': [20, 25, 15, 30, 18, 22, 16, 35, 12]
}

def compute_ranks(values):
    """Calculate ranks with proper tie handling"""
    sorted_pairs = sorted((val, idx) for idx, val in enumerate(values))
    ranks = [0.0] * len(values)
    n = len(values)
    i = 0
    
    while i < n:
        j = i
        while j < n and sorted_pairs[j][0] == sorted_pairs[i][0]:
            j += 1
        avg_rank = (i + 1 + j) / 2.0  # 1-based indexing
        for k in range(i, j):
            original_idx = sorted_pairs[k][1]
            ranks[original_idx] = avg_rank
        i = j
    return ranks

def print_analysis(x, y, x_name, y_name):
    """Perform full Spearman analysis with detailed output"""
    # Get ranks
    x_rank = compute_ranks(x)
    y_rank = compute_ranks(y)
    
    print(f"\n{'='*60}")
    print(f"Analysis of {x_name} vs {y_name}")
    print(f"{'='*60}")
    
    # Print raw data
    print("\nOriginal Data:")
    print(f"{x_name:>15} | {y_name:>15}")
    print("-"*35)
    for a, b in zip(x, y):
        print(f"{a:15} | {b:15}")
    
    # Print ranks
    print("\nRanks with Tie Handling:")
    print(f"{x_name+' Rank':>15} | {y_name+' Rank':>15} | {'Difference (d)':>15} | {'d²':>15}")
    print("-"*75)
    d_squared_total = 0
    for i in range(len(x)):
        d = x_rank[i] - y_rank[i]
        d_sq = d ** 2
        d_squared_total += d_sq
        print(f"{x_rank[i]:15.2f} | {y_rank[i]:15.2f} | {d:15.2f} | {d_sq:15.2f}")
    
    # Calculate Spearman's rho
    n = len(x)
    numerator = 6 * d_squared_total
    denominator = n * (n**2 - 1)
    rho = 1 - numerator/denominator
    
    print("\nIntermediate Calculations:")
    print(f"Sum of d² (Σd²) = {d_squared_total:.2f}")
    print(f"n(n² - 1) = {n}*({n}²-1) = {denominator}")
    print(f"Spearman's ρ = 1 - (6*{d_squared_total:.2f})/{denominator} = {rho:.3f}")
    
    return rho

# Perform analyses
print("STARTING ANALYSIS...")
car_age_rho = print_analysis(data['Car Age'], data['Price'], "Car Age", "Price")
owners_rho = print_analysis(data['Previous Owners'], data['Price'], "Previous Owners", "Price")

# Final conclusion
print("\n\nFINAL CONCLUSION:")
print(f"{'Car Age vs Price Correlation:':<35} ρ = {car_age_rho:.3f}")
print(f"{'Previous Owners vs Price Correlation:':<35} ρ = {owners_rho:.3f}")

if abs(car_age_rho) > abs(owners_rho):
    conclusion = ("Car Age is the stronger predictor of Price. There is a perfect negative correlation "
                  "(ρ = -1.0), meaning car price decreases consistently as vehicles get older.")
else:
    conclusion = ("Previous Owners is the stronger predictor of Price. There is a strong negative correlation "
                  "(ρ = -0.750), meaning prices tend to decrease with more previous owners, but not perfectly.")

print(f"\n{conclusion}")
print("Note: Negative correlation means as the predictor increases, price decreases")
```

```python
# Dataset
data = {
    'Car Age': [5, 3, 8, 2, 6, 4, 7, 1, 9],
    'Previous Owners': [1, 2, 3, 1, 2, 1, 3, 1, 4],
    'Distance Driven': [50, 30, 80, 20, 60, 40, 70, 10, 90],
    'Price': [20, 25, 15, 30, 18, 22, 16, 35, 12]
}

def rank_data(arr):
    """Calculate ranks with tie handling using average ranking"""
    sorted_arr = sorted((val, idx) for idx, val in enumerate(arr))
    ranks = [0] * len(arr)
    
    print("\nRanking Process:")
    for i, (val, idx) in enumerate(sorted_arr):
        # Find all elements with same value
        duplicates = [j for j, (v, _) in enumerate(sorted_arr) if v == val]
        avg_rank = sum(duplicates)/len(duplicates) + 1  # +1 for 1-based indexing
        for j in duplicates:
            ranks[sorted_arr[j][1]] = avg_rank
        print(f"Value {val} at positions {[sorted_arr[d][1] for d in duplicates]} gets rank {avg_rank:.1f}")
    
    return ranks

def spearman_rank(x, y, x_name, y_name):
    """Calculate Spearman's rank correlation coefficient with detailed output"""
    print(f"\n{'='*40}\nCalculating Spearman for {x_name} vs {y_name}\n{'='*40}")
    
    # Step 1: Rank the data
    x_rank = rank_data(x)
    y_rank = rank_data(y)
    
    print(f"\n{x_name} Ranks: {[round(r,1) for r in x_rank]}")
    print(f"{y_name} Ranks: {[round(r,1) for r in y_rank]}")
    
    # Step 2: Calculate differences
    d = [xr - yr for xr, yr in zip(x_rank, y_rank)]
    d_sq = [di**2 for di in d]
    
    print("\nPair Calculations:")
    print(f"{'Index':<5} | {x_name:<7} | {y_name:<7} | {'d':<6} | {'d²':<6}")
    print("-"*40)
    for i in range(len(x)):
        print(f"{i+1:<5} | {x_rank[i]:<7.1f} | {y_rank[i]:<7.1f} | {d[i]:<6.1f} | {d_sq[i]:<6.1f}")
    
    # Step 3: Calculate components
    sum_d_sq = sum(d_sq)
    n = len(x)
    numerator = 6 * sum_d_sq
    denominator = n * (n**2 - 1)
    rho = 1 - numerator/denominator
    
    print("\nFinal Calculation:")
    print(f"Σd² = {sum_d_sq}")
    print(f"n(n²-1) = {n}*({n}²-1) = {denominator}")
    print(f"rho = 1 - (6*{sum_d_sq})/{denominator} = {rho:.3f}")
    
    return rho

def kendalls_tau(x, y, x_name, y_name):
    """Calculate Kendall's Tau with detailed pair analysis"""
    print(f"\n{'='*40}\nCalculating Kendall's Tau for {x_name} vs {y_name}\n{'='*40}")
    
    n = len(x)
    pairs = []
    C = 0  # Concordant pairs
    D = 0  # Discordant pairs
    
    print("\nPair-wise Analysis:")
    for i in range(n):
        for j in range(i+1, n):
            x_dir = x[i] - x[j]
            y_dir = y[i] - y[j]
            
            if x_dir * y_dir > 0:
                pairs.append(('C', i+1, j+1))
                C += 1
            elif x_dir * y_dir < 0:
                pairs.append(('D', i+1, j+1))
                D += 1
            else:
                pairs.append(('T', i+1, j+1))
    
    # Print first 5 pairs for demonstration
    for p in pairs[:5]:
        print(f"Pair ({p[1]},{p[2]}): {p[0]}")
    print(f"... Total {len(pairs)} pairs analyzed")
    
    total_pairs = n*(n-1)/2
    tau = (C - D) / total_pairs
    
    print("\nCount Summary:")
    print(f"Concordant pairs (C): {C}")
    print(f"Discordant pairs (D): {D}")
    print(f"Total pairs: {total_pairs}")
    print(f"Kendall's Tau = (C-D)/Total = ({C}-{D})/{total_pairs} = {tau:.3f}")
    
    return tau

# # Problem 1: Compare Car Age vs Price and Previous Owners vs Price
# print("="*60)
# print("PROBLEM 1: Predictor Comparison using Spearman")
# print("="*60)

# # Car Age vs Price
# car_age_price = spearman_rank(data['Car Age'], data['Price'], "Car Age", "Price")

# # Previous Owners vs Price
# owners_price = spearman_rank(data['Previous Owners'], data['Price'], "Previous Owners", "Price")

# print("\nComparison Results:")
# print(f"Car Age vs Price: ρ = {car_age_price:.3f}")
# print(f"Previous Owners vs Price: ρ = {owners_price:.3f}")

# if abs(car_age_price) > abs(owners_price):
#     conclusion = "Car Age is a better predictor of Price"
# else:
#     conclusion = "Previous Owners is a better predictor of Price"

# print(f"\nConclusion: {conclusion} (higher absolute correlation)")

# Problem 2: Kendall's Tau between Distance Driven and Price
print("\n" + "="*60)
print("PROBLEM 2: Kendall's Tau Calculation")
print("="*60)

kendall_result = kendalls_tau(data['Distance Driven'], data['Price'], "Distance Driven", "Price")
print(f"\nFinal Kendall's Tau: τ = {kendall_result:.3f}")
```
